```julia
push!(LOAD_PATH, joinpath(Pkg.dir("RGBDSegmentation"), "src","Evaluation"))
using TestImages, ImageSegmentation, Plots, Colors,  Images, Evaluation

pyplot()
# visualize segmentation by creating an image with each label replaced by a random color
function get_random_color(seed)
    srand(seed)
    rand(RGB{N0f8})
end


img = Gray.(testimage("house"));
segments1 = felzenszwalb(img, 300, 100);
segments2 = felzenszwalb(img, 150, 200);


labels_1 = labels_map(segments1)
labels_2 = labels_map(segments2)
b = Evaluation.evaluate(FBoundary(3), labels_1, labels_2)
```

```julia
labels_1 = labels_map(segments1)
labels_2 = labels_map(segments2)

criteria = [Precision(), 
            FMeasure(), 
            SegmentationCovering(), 
            VariationOfInformation(true), 
            RandIndex(), 
            FMeasureRegions(), 
            PRObjectsAndParts(0.9, 0.25 , 0.1),
            BoundaryDisplacementError(),
            FBoundary(4)]
[Evaluation.evaluate(c, labels_2, labels_1) for c in criteria]
```
